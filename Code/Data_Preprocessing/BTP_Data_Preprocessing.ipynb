```python
df_noa = pd.read_excel('/content/NOA.xlsx')
df_pdo = pd.read_excel('/content/PDO.xlsx')
df_nino = pd.read_csv('/content/Nino_3.4.csv')
df_soi = pd.read_excel('/content/SOI.xlsx')
df_independent_variables = pd.read_excel('/content/Independent_variables.xlsx')
```

```python
import pandas as pd

df_noa_new = df_noa

df_noa_new = pd.melt(df_noa_new, id_vars=['Year'], var_name='Month', value_name='Value')
df_noa_new['Date'] = df_noa_new['Year'].astype(str) + '-' + df_noa_new['Month']
df_noa_new = df_noa_new.drop(columns=['Year', 'Month'])
df_noa_new = df_noa_new.set_index('Date')

df_noa_new.index = pd.to_datetime(df_noa_new.index, format='%Y-%b')
df_noa_new.index = df_noa_new.index + pd.DateOffset(day=1)

df_noa_new = df_noa_new.sort_index()

df_noa_new.to_excel('NOA_FINAL.xlsx')
```

```python
import pandas as pd

df_pdo_new = df_pdo

df_pdo_new = pd.melt(df_pdo_new, id_vars=['Year'], var_name='Month', value_name='Value')
df_pdo_new['Date'] = df_pdo_new['Year'].astype(str) + '-' + df_pdo_new['Month']
df_pdo_new = df_pdo_new.drop(columns=['Year', 'Month'])
df_pdo_new = df_pdo_new.set_index('Date')

df_pdo_new.index = pd.to_datetime(df_pdo_new.index, format='%Y-%b')
df_pdo_new.index = df_pdo_new.index + pd.DateOffset(day=1)

df_pdo_new = df_pdo_new.sort_index()

df_pdo_new.to_excel('PDO_FINAL.xlsx')
```

```python
import pandas as pd

df_soi_new = df_soi

df_soi_new = pd.melt(df_soi_new, id_vars=['YEAR'], var_name='Month', value_name='Value')
df_soi_new['Date'] = df_soi_new['YEAR'].astype(str) + '-' + df_soi_new['Month']
df_soi_new = df_soi_new.drop(columns=['YEAR', 'Month'])
df_soi_new = df_soi_new.set_index('Date')

df_soi_new.index = pd.to_datetime(df_soi_new.index, format='%Y-%b')
df_soi_new.index = df_soi_new.index + pd.DateOffset(day=1)

df_soi_new = df_soi_new.sort_index()

df_soi_new.to_excel('SOI_FINAL.xlsx')
```

```python
df_nino_new = df_nino

df_nino_new['Date'] = pd.to_datetime(df_nino_new['YR'].astype(str) + '-' + df_nino_new['MON'].astype(str), format='%Y-%m')

df_nino_new.set_index('Date', inplace=True)
df_nino_new.drop(['YR', 'MON'], axis=1, inplace=True)

df_nino_new.to_excel('NINO_FINAL.xlsx')
```

```python
df_independent_variables.drop(['longitude','latitude','u10','v10','t2m','sp'], axis=1, inplace=True)
df_independent_variables.set_index('time', inplace=True)
df_independent_variables.to_excel('TP.xlsx')
```
