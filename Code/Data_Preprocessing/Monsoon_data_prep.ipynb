```python
df = pd.read_excel('/content/Data.xlsx')
df.head()
```

```python
import pandas as pd

# Assuming your existing dataset is named 'df'
# Extracting monsoon data
monsoon_data = df[(df['Date'].dt.month >= 6) & (df['Date'].dt.month <= 9)]

# Calculating monsoon totals for each year
monsoon_totals = monsoon_data.groupby(monsoon_data['Date'].dt.year)['TP'].sum().reset_index()
monsoon_totals = monsoon_totals.rename(columns={'TP': 'MonsoonTotalPrecipitation'})

# Creating a DataFrame with one row per year
years_range = range(1951, 2024)  # Adjust the range based on your dataset
monsoon_dataframe = pd.DataFrame({'Year': years_range})
monsoon_dataframe['MonsoonTotalPrecipitation'] = monsoon_dataframe['Year'].map(monsoon_totals.set_index('Date')['MonsoonTotalPrecipitation'])

# Displaying the new DataFrame
print(monsoon_dataframe)
```

```python
monsoon_dataframe.to_excel('Monsoon_data.xlsx')
```

```python
import pandas as pd

# Assuming your DataFrame is named 'monsoon_dataframe'
# Calculate standard deviation and mean
std_deviation = monsoon_dataframe['MonsoonTotalPrecipitation'].std()
mean_value = monsoon_dataframe['MonsoonTotalPrecipitation'].mean()

# Display the results
print(f"Standard Deviation: {std_deviation}")
print(f"Mean: {mean_value}")
```
