```python
import pandas as pd
import numpy as np
```

```python
df_dmi = pd.read_excel('/content/DMI.xlsx')
df_nino = pd.read_excel('/content/NINO_FINAL.xlsx')
df_noa = pd.read_excel('/content/NOA_FINAL.xlsx')
df_pdo = pd.read_excel('/content/PDO_FINAL.xlsx')
df_soi = pd.read_excel('/content/SOI_FINAL.xlsx')
df_tp = pd.read_excel('/content/TP.xlsx')

df_dmi = df_dmi.dropna()
df_nino = df_nino.dropna()
df_noa = df_noa.dropna()
df_pdo = df_pdo.dropna()
df_soi = df_soi.dropna()
df_tp = df_tp.dropna()
```

```python
df_dmi['time'] = pd.to_datetime(df_dmi['time'])

start_date = pd.to_datetime('1940-01-01')
end_date = pd.to_datetime('1950-12-01')

df_dmi = df_dmi[(df_dmi['time'] < start_date) | (df_dmi['time'] > end_date)]
```

```python
df_nino['date'] = pd.to_datetime(df_nino['date'])

start_date = pd.to_datetime('1940-01-01')
end_date = pd.to_datetime('1950-12-01')

df_nino = df_nino[(df_nino['date'] < start_date) | (df_nino['date'] > end_date)]
```

```python
df_noa['date'] = pd.to_datetime(df_noa['date'])

start_date = pd.to_datetime('1940-01-01')
end_date = pd.to_datetime('1950-12-01')

df_noa = df_noa[(df_noa['date'] < start_date) | (df_noa['date'] > end_date)]
```

```python
df_pdo['date'] = pd.to_datetime(df_pdo['date'])

start_date = pd.to_datetime('1940-01-01')
end_date = pd.to_datetime('1950-12-01')

df_pdo = df_pdo[(df_pdo['date'] < start_date) | (df_pdo['date'] > end_date)]
```

```python
df_soi['date'] = pd.to_datetime(df_soi['date'])

start_date = pd.to_datetime('1940-01-01')
end_date = pd.to_datetime('1950-12-01')

df_soi = df_soi[(df_soi['date'] < start_date) | (df_soi['date'] > end_date)]
```

```python
df_tp['date'] = pd.to_datetime(df_tp['date'])

start_date = pd.to_datetime('1940-01-01')
end_date = pd.to_datetime('1950-12-01')

df_tp = df_tp[(df_tp['date'] < start_date) | (df_tp['date'] > end_date)]
```

```python
df_dmi = df_dmi.rename(columns={'value':'DMI'})
```

```python
df_nino = df_nino.rename(columns={'value':'NINO'})
```

```python
df_noa = df_noa.rename(columns={'value':'NOA'})
```

```python
df_pdo = df_pdo.rename(columns={'value':'PDO'})
```

```python
df_soi = df_soi.rename(columns={'value':'SOI'})
```

```python
df_tp = df_tp.rename(columns={'tp':'TP'})
```

```python
merged_df = pd.concat([df_dmi, df_nino, df_noa, df_pdo, df_soi,df_tp], axis=1)
merged_df
```

```python
merged_df.to_excel('Data.xlsx')
```
