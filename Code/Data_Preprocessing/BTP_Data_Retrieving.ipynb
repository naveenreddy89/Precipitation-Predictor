```python
import netCDF4 as nc
import xarray as xr
```

```python
ds_eastern_sst1 = xr.open_dataset('/content/Eastern_SST.nc')

target_latitude = 5.25
target_longitude = 99.75

ds_eastern_sst2 = ds_eastern_sst1.sel(latitude = target_latitude, longitude = target_longitude, method = 'nearest')

ds_eastern_sst3 = ds_eastern_sst2.reduce(np.nansum, dim='expver',keep_attrs=True)

df_eastern_sst = ds_eastern_sst3.to_dataframe()
df_eastern_sst
```

```python
df_eastern_sst.to_excel('Eastern_SST.xlsx', index=True)
```

```python
ds_western_sst1 = xr.open_dataset('/content/Western_SST.nc')

target_latitude = 10.25
target_longitude = 59.75

ds_western_sst2 = ds_western_sst1.sel(latitude = target_latitude, longitude = target_longitude, method = 'nearest')

ds_western_sst3 = ds_western_sst2.reduce(np.nansum, dim='expver',keep_attrs=True)

df_western_sst = ds_western_sst3.to_dataframe()
df_western_sst
```

```python
df_western_sst.to_excel('Western_SST.xlsx', index=True)
```

```python
import csv

soi_data = []

with open('/content/SOI_text_data.txt', 'r') as file:
    lines = file.readlines()
    for line in lines:
        values = line.strip().split()
        soi_data.append(values)

with open('SOI.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    for row in soi_data:
        csvwriter.writerow(row)

soi_df = pd.read_csv('/content/SOI.csv')

soi_df.to_excel("SOI.xlsx", index=False)
```
