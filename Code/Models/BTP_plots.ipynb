```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('/content/RMSE_values.xlsx')
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))

plt.plot(df['Batch size'].iloc[[0, 1, 2]], df['Rmse'].iloc[[0, 1, 2]], color='#FF0000', marker='^', linestyle='-', linewidth=0.8, label='RMSE')
plt.plot(df['Batch size'].iloc[[0, 1, 2]], df['R'].iloc[[0, 1, 2]], color='#FF0000', marker='^', linestyle='-', linewidth=0.8, label='RMSE')

plt.title('Batch size vs. Root Mean Square Error (RMSE)', weight='light')
plt.xlabel('Batch size', weight='light')
plt.ylabel('Root Mean Square Error (RMSE)', weight='light')

plt.ylim(0.25, 0.32)
plt.grid(False)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))

plt.plot(df['Window Length'].iloc[[0, 3, 4, 5, 6, 7]], df['Rmse'].iloc[[0, 3, 4, 5, 6, 7]], color='r', marker='^', linestyle='-', linewidth=1.2, label='RMSE')
plt.plot(df['Window Length'].iloc[[0, 3, 4, 5, 6, 7]], df['R'].iloc[[0, 3, 4, 5, 6, 7]], color='r', marker='^', linestyle='-', linewidth=1.2, label='RMSE')

plt.title('Window Length vs. Root Mean Square Error (RMSE)')
plt.xlabel('Window Length')
plt.ylabel('Root Mean Square Error (RMSE)')

plt.ylim(0.23, 0.29)
plt.grid(False)

plt.xticks(df['Window Length'].iloc[[0, 3, 4, 5, 6, 7]], [12, 24, 36, 48, 60, 72])

plt.legend()
plt.show()
```
